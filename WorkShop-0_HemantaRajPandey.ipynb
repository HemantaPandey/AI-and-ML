```python
# Unit Conversion Program in Jupyter Notebook

def length_conversion(value, unit):
    if unit == "m":
        return value * 3.28084  # Meters to feet
    elif unit == "ft":
        return value / 3.28084  # Feet to meters
    else:
        raise ValueError("Unsupported length unit. Use 'm' or 'ft'.")

def weight_conversion(value, unit):
    if unit == "kg":
        return value * 2.20462  # Kg to lbs
    elif unit == "lbs":
        return value / 2.20462  # Lbs to kg
    else:
        raise ValueError("Unsupported weight unit. Use 'kg' or 'lbs'.")

def volume_conversion(value, unit):
    if unit == "L":
        return value * 0.264172  # Liters to gallons
    elif unit == "gal":
        return value / 0.264172  # Gallons to liters
    else:
        raise ValueError("Unsupported volume unit. Use 'L' or 'gal'.")

def main():
    print("Unit Conversion Program")
    print("1. Length (m <-> ft)")
    print("2. Weight (kg <-> lbs)")
    print("3. Volume (L <-> gal)")
    
    try:
        choice = int(input("Choose conversion type (1-3): "))
        value = float(input("Enter the value to convert: "))
        
        if choice == 1:
            unit = input("Enter unit (m or ft): ").strip().lower()
            converted_value = length_conversion(value, unit)
            new_unit = "ft" if unit == "m" else "m"
        
        elif choice == 2:
            unit = input("Enter unit (kg or lbs): ").strip().lower()
            converted_value = weight_conversion(value, unit)
            new_unit = "lbs" if unit == "kg" else "kg"
        
        elif choice == 3:
            unit = input("Enter unit (L or gal): ").strip().lower()
            converted_value = volume_conversion(value, unit)
            new_unit = "gal" if unit == "L" else "L"
        
        else:
            print("Invalid choice! Please select 1, 2, or 3.")
            return
        
        print(f"{value} {unit} is equal to {converted_value:.2f} {new_unit}")
    
    except ValueError as e:
        print(f"Error: {e}")


main()
```

```python
# Mathematical Operations Program in Jupyter Notebook

def calculate_sum(numbers):
    return sum(numbers)

def calculate_average(numbers):
    return sum(numbers) / len(numbers) if numbers else 0

def find_maximum(numbers):
    return max(numbers)

def find_minimum(numbers):
    return min(numbers)

def get_number_list():
    while True:
        try:
            numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))
            if not numbers:
                raise ValueError("List cannot be empty. Please enter at least one number.")
            return numbers
        except ValueError:
            print("Invalid input! Please enter numbers only.")

def main():
    print("Mathematical Operations Program")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")
    
    try:
        choice = int(input("Choose an operation (1-4): "))
        numbers = get_number_list()
        
        if choice == 1:
            result = calculate_sum(numbers)
            operation = "Sum"
        elif choice == 2:
            result = calculate_average(numbers)
            operation = "Average"
        elif choice == 3:
            result = find_maximum(numbers)
            operation = "Maximum"
        elif choice == 4:
            result = find_minimum(numbers)
            operation = "Minimum"
        else:
            print("Invalid choice! Please select 1, 2, 3, or 4.")
            return
        
        print(f"{operation} of {numbers} is: {result:.2f}")
    
    except ValueError:
        print("Error: Invalid input! Please enter a valid number.")


main()
```

```python
# list_operations.py

# 1. Flatten a Nested List
def flatten(lst):
    """Flattens a nested list into a single dimension."""
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item)) 
        else:
            flat_list.append(item)
    return flat_list


# 2. Accessing Nested List Elements
def access_nested_element(lst, indices):
    """Extracts an element from a nested list based on a list of indices."""
    for index in indices:
        lst = lst[index]
    return lst


# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """Calculates the sum of all elements in a nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# 4. Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """Removes all occurrences of a specific element from a nested list."""
    if isinstance(lst, list):
        return [remove_element(i, elem) for i in lst if i != elem]
    return lst


# 5. Find the Maximum Element in a Nested List
def find_max(lst):
    """Finds the maximum element in a nested list."""
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))  
        else:
            max_val = max(max_val, item)
    return max_val


# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """Counts how many times a specific element appears in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  
        elif item == elem:
            count += 1
    return count


# 7. Flatten a List of Lists of Lists
def deep_flatten(lst):
    """Flattens a deeply nested list into a single dimension."""
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item)) 
        else:
            flat_list.append(item)
    return flat_list


# 8. Nested List Average
def average_nested(lst):
    """Calculates the average of all elements in a nested list."""
    total = sum_nested(lst)
    count = sum(1 for _ in flatten(lst))
    return total / count if count > 0 else 0



# Test Cases for Each Function
print(flatten([[1, 2], [3, 4], [5]])) 
print(access_nested_element([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [1, 2])) 
print(sum_nested([[1, 2], [3, [4, 5]], 6])) 
print(remove_element([[1, 2], [3, 2], [4, 5]], 2))  
print(find_max([[1, 2], [3, [4, 5]], 6])) 
print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))  
print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])) 
print(average_nested([[1, 2], [3, 4], [5, 6]])) 
```

```python
import numpy as np

# Problem 1: Prove A * A^(-1) = I
A = np.array([[3, 4], [7, 8]])
A_inv = np.linalg.inv(A)  # Compute the inverse of A
I = np.dot(A, A_inv)  # Multiply A and A_inv to get the identity matrix

print("Problem 1: A * A^(-1) = I")
print("A * A^(-1) = \n", I)
print("\n")

# Problem 2: Prove AB != BA
B = np.array([[5, 3], [2, 1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

print("Problem 2: Prove AB != BA")
print("AB = \n", AB)
print("BA = \n", BA)
print("\n")

# Problem 3: Prove (AB)^T = B^T * A^T
AB_T = np.transpose(np.dot(A, B))
BT_AT = np.dot(np.transpose(B), np.transpose(A))

print("Problem 3: Prove (AB)^T = B^T * A^T")
print("(AB)^T = \n", AB_T)
print("B^T * A^T = \n", BT_AT)
print("\n")

# Problem 4: Solve the system of Linear equations using Inverse Methods
# The system of equations is:
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

# Represent the system in matrix form AX = B
A_system = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_system = np.array([-1, -3, 9])

# Solve for X using the inverse method: X = A_inv * B
A_inv_system = np.linalg.inv(A_system)  
X_solution = np.dot(A_inv_system, B_system)  

print("Problem 4: Solve the system of Linear equations using Inverse Methods")
print("Solution for x, y, z: ", X_solution)
print("\n")

# Problem 5: Solve the system using np.linalg.inv
X_solution_linalg = np.linalg.solve(A_system, B_system)  

print("Problem 5: Solve the system using np.linalg.inv")
print("Solution using np.linalg.solve: ", X_solution_linalg)
```

```python
import time
import numpy as np

# 1. Element-wise Addition
# Using Python Lists
list1 = [i for i in range(1000000)]
list2 = [i for i in range(1000000)]
start_time = time.time()
list_sum = [list1[i] + list2[i] for i in range(len(list1))]
python_list_add_time = time.time() - start_time
print(f"Element-wise addition using Python lists took: {python_list_add_time:.6f} seconds")

# Using NumPy Arrays
np_array1 = np.arange(1000000)
np_array2 = np.arange(1000000)
start_time = time.time()
np_sum = np_array1 + np_array2
numpy_add_time = time.time() - start_time
print(f"Element-wise addition using NumPy arrays took: {numpy_add_time:.6f} seconds")

# 2. Element-wise Multiplication
# Using Python Lists
start_time = time.time()
list_multiplication = [list1[i] * list2[i] for i in range(len(list1))]
python_list_multiply_time = time.time() - start_time
print(f"Element-wise multiplication using Python lists took: {python_list_multiply_time:.6f} seconds")

# Using NumPy Arrays
start_time = time.time()
np_multiplication = np_array1 * np_array2
numpy_multiply_time = time.time() - start_time
print(f"Element-wise multiplication using NumPy arrays took: {numpy_multiply_time:.6f} seconds")

# 3. Dot Product
# Using Python Lists
start_time = time.time()
dot_product_python = sum(list1[i] * list2[i] for i in range(len(list1)))
python_dot_time = time.time() - start_time
print(f"Dot product using Python lists took: {python_dot_time:.6f} seconds")

# Using NumPy Arrays
start_time = time.time()
dot_product_numpy = np.dot(np_array1, np_array2)
numpy_dot_time = time.time() - start_time
print(f"Dot product using NumPy arrays took: {numpy_dot_time:.6f} seconds")

# 4. Matrix Multiplication
# Using Python Lists
matrix1 = [[i + j for j in range(1000)] for i in range(1000)]
matrix2 = [[i - j for j in range(1000)] for i in range(1000)]
start_time = time.time()
matrix_multiply_python = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(1000)) for j in range(1000)] for i in range(1000)]
python_matrix_multiply_time = time.time() - start_time
print(f"Matrix multiplication using Python lists took: {python_matrix_multiply_time:.6f} seconds")

# Using NumPy Arrays
np_matrix1 = np.arange(1000000).reshape(1000, 1000)
np_matrix2 = np.arange(1000000, 0, -1).reshape(1000, 1000)
start_time = time.time()
matrix_multiply_numpy = np.dot(np_matrix1, np_matrix2)
numpy_matrix_multiply_time = time.time() - start_time
print(f"Matrix multiplication using NumPy arrays took: {numpy_matrix_multiply_time:.6f} seconds")
```
